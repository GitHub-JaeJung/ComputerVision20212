# 1.4.7 Canny Edge Detection

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('messi5.jpg', 0)
blur = cv2.GaussianBlur(img,(5,5),0)
edges50 = cv2.Canny(blur, 50, 200)
edges100 = cv2.Canny(blur, 100, 200)
edges180 = cv2.Canny(blur, 180, 200)
```

```python
plt.rcParams["figure.figsize"] = (20, 10)
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2),plt.imshow(edges50,cmap = 'gray')
plt.title('Edge50 Image'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3),plt.imshow(edges100,cmap = 'gray')
plt.title('Edge100 Image'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,4),plt.imshow(edges180,cmap = 'gray')
plt.title('Edge180 Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
